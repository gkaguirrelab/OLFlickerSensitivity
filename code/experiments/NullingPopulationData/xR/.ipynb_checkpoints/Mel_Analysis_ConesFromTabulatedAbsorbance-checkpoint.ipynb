```r
library(lmerTest)
library(ggplot2)
options(width=150)
```

```r
mydata = read.csv("/Users/mspits/Dropbox (Aguirre-Brainard Lab)/MELA_analysis/NullingPopulationData/xLMSTabulatedAbsorbance/Mel_AllNullingContrast_LMSFromTabulatedAbsorbance.csv")
mydata$Run <- factor(mydata$Run)
mydata$Run <- gsub('1', 'Run 1', mydata$Run)
mydata$Run <- gsub('2', 'Run 2', mydata$Run)
mydata$Arm <- factor(mydata$Arm)
mydata$Arm <- gsub('1', 'Positive', mydata$Arm)
mydata$Arm <- gsub('2', 'Negative', mydata$Arm)
mydata$Group <- factor(mydata$Group)
mydata$Group <- gsub('NullingPopulationData18To24', '18-24 yrs', mydata$Group)
mydata$Group <- gsub('NullingPopulationDataFork2', '25-40 yrs', mydata$Group)
```

```r
testLMS <- lmer(LMS.null ~ (1 | Subject) + Group + Arm + Run, data=mydata)
lsmeans(testLMS)
difflsmeans(testLMS)
fileName <- '/Users/mspits/Dropbox (Aguirre-Brainard Lab)/MELA_analysis/NullingPopulationData/xLMSTabulatedAbsorbance//Mel_LMSNull_ANOVA_LMSFromTabulatedAbsorbance.csv'
capture.output(lsmeans(testLMS), file=fileName)
capture.output(difflsmeans(testLMS), file=fileName, append=TRUE)
```

```r
testLMinusM <- lmer(L.M.null ~ (1 | Subject) + Group + Arm + Run, data=mydata)
lsmeans(testLMinusM)
difflsmeans(testLMinusM)
fileName <- '/Users/mspits/Dropbox (Aguirre-Brainard Lab)/MELA_analysis/NullingPopulationData/xLMSTabulatedAbsorbance/Mel_LMinusMNull_ANOVA_LMSFromTabulatedAbsorbance.csv'
capture.output(lsmeans(testLMinusM), file=fileName)
capture.output(difflsmeans(testLMinusM), file=fileName, append=TRUE)
```

```r
testS <- lmer(S.null ~ (1 | Subject) + Group + Arm + Run, data=mydata)
lsmeans(testS)
difflsmeans(testS)
fileName <- '/Users/mspits/Dropbox (Aguirre-Brainard Lab)/MELA_analysis/NullingPopulationData/xLMSTabulatedAbsorbance/Mel_SNull_ANOVA_LMSFromTabulatedAbsorbance.csv'
capture.output(lsmeans(testS), file=fileName)
capture.output(difflsmeans(testS), file=fileName, append=TRUE)
```

```r
b1 <- plot(lsmeans(testLMS))
b1 + ggtitle("Mel directed / LMS null") + coord_cartesian(ylim=c(-0.08, 0.08)) + ylab('Contrast') + xlab('Groups') + theme(legend.position="none")
ggsave(file='/Users/mspits/Dropbox (Aguirre-Brainard Lab)/MELA_analysis/NullingPopulationData/xLMSTabulatedAbsorbance/Mel_LMSNull_ANOVA_LMSFromTabulatedAbsorbance.pdf')
b2 <- plot(lsmeans(testLMinusM))
b2 + ggtitle("Mel directed / L-M null") + coord_cartesian(ylim=c(-0.08, 0.08)) + ylab('Contrast') + xlab('Groups') + theme(legend.position="none")
ggsave(file='/Users/mspits/Dropbox (Aguirre-Brainard Lab)/MELA_analysis/NullingPopulationData/xLMSTabulatedAbsorbance/Mel_LMinusMNull_ANOVA_LMSFromTabulatedAbsorbance.pdf')
b3 <- plot(lsmeans(testS))
b3 + ggtitle("Mel directed / S null") + coord_cartesian(ylim=c(-0.08, 0.08)) + ylab('Contrast') + xlab('Groups') + theme(legend.position="none")
ggsave(file='/Users/mspits/Dropbox (Aguirre-Brainard Lab)/MELA_analysis/NullingPopulationData/xLMSTabulatedAbsorbance/Mel_SNull_ANOVA_LMSFromTabulatedAbsorbance.pdf')
```
